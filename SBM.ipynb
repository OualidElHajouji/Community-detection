```python
import numpy as np
import matplotlib.pyplot as plt
import igraph




n = 100
cin = 100
cout= 10
k = 2
def simulate(B,Classes): 
    #Given a matrix B indicating probabilities of links between two elements of a class, simulation of the graph
    U = np.random.rand(n,n)
    Matrix = np.array([[B[Classes[i],Classes[j]] for i in range(n)] for j in range(n)])
    A = (U<=Matrix)
    return A


B = (cin/n-cout/n)*np.identity(k) + (cout/n)*np.ones((k,k))
Classes = np.random.binomial(size = n, n = 1, p = 0.5)

A = simulate(B, Classes)

G = igraph.Graph.Adjacency(A.tolist())

print(1)
layout = G.layout_lgl()
print(2)
igraph.plot(G, layout = layout)
```
