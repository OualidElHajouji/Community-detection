```python
import numpy as np
import matplotlib.pyplot as plt
import igraph




n = 1000
cin = 100
cout= 10
k = 2
def simulate(B,Classes): 
    #Given a matrix B indicating probabilities of links between two elements of a class, simulation of the graph
    U = np.random.rand(n,n)
    U = (U + U.T)/2
    Matrix = np.array([[B[Classes[i],Classes[j]]*(i !=j) for i in range(n)] for j in range(n)])
    A = (U<=Matrix)
    return A

color_list = ['blue', 'red', 'pink', 'yellow', 'green']
B = (cin/n-cout/n)*np.identity(k) + (cout/n)*np.ones((k,k))
p = k*[1/k]
Classes = np.random.choice(range(k), size = n, p = p)

A = simulate(B, Classes)

G = igraph.Graph.Adjacency(A.tolist())
G.to_undirected()
layout = G.layout_lgl()
igraph.plot(G, layout = layout, vertex_color = [color_list[classe] for classe in Classes])
```
